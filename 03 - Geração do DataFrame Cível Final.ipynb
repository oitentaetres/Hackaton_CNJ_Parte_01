# Importação das bibliotecas:

```python
import ast
import pandas
```

# Leitura do Dataframe Preliminar gerado na etapa anterior:

```python
dfPreliminar = pandas.read_csv('dfPreliminar.csv', low_memory = False)

dfPreliminar.head()
```

# Escolha da classe processual:

```python
# 7 = Processo Comum Cível:

classeProcessual = 7
```

# Geração do Dataframe Final:

```python
data = []

necessidadeNumeracao = False

for i in range(dfPreliminar.shape[0]):    
    
    if (dfPreliminar.loc[i, 'dadosBasicos.classeProcessual'] == classeProcessual):
        
        numero = dfPreliminar.loc[i, 'dadosBasicos.numero']

        ajuizamento = dfPreliminar.loc[i, 'dadosBasicos.dataAjuizamento']

        grau = dfPreliminar.loc[i, 'grau']

        tribunal = dfPreliminar.loc[i, 'siglaTribunal']

        classe = dfPreliminar.loc[i, 'dadosBasicos.classeProcessual']

        listaMovimentos = dfPreliminar.loc[i, 'movimento']

        if (type(listaMovimentos) == str):

            listaMovimentos = ast.literal_eval(listaMovimentos) 
            
            for j in range(len(listaMovimentos)):

                movimento = listaMovimentos[j]

                try:
                    
                    identificadorMovimento = movimento['identificadorMovimento']

                except:
                    
                    identificadorMovimento = 0
                    
                    necessidadeNumeracao = True
                
                dataHora = movimento['dataHora']

                try:

                    codigoNacional = movimento['movimentoNacional']['codigoNacional']

                except:

                    codigoNacional = None

                try:

                    codigoLocal = movimento['movimentoLocal']['codigoMovimento']

                    codigoPaiNacional = movimento['movimentoLocal']['codigoPaiNacional']

                except:

                    codigoLocal = None

                    codigoPaiNacional = None

                data.append([
                    numero,
                    ajuizamento,
                    grau,
                    tribunal,
                    classe,
                    identificadorMovimento,
                    dataHora,
                    codigoNacional,
                    codigoLocal,
                    codigoPaiNacional
                ])
                
        else:
            
            data.append([
                numero,
                ajuizamento,
                grau,
                tribunal,
                classe,
                None,
                None,
                None,
                None,
                None
            ])
            
            
                     
colunas = [
    'numero', 
    'ajuizamento', 
    'grau',
    'tribunal',
    'classeProcessual',
    'identificadorMovimento',  
    'dataHora',
    'codigoNacional',
    'codigoLocal',
    'codigoPaiNacional'
]
           
        
dfFinalCivel = pandas.DataFrame(data, columns = colunas)

dfFinalCivel.sort_values(['numero', 'dataHora'], inplace = True, ignore_index = True)
            
dfFinalCivel.head()
```

# Exportação do Dataframe Final em formato CSV:

```python
dfFinalCivel.to_csv("dfFinalCivel.csv", index = False)
```
