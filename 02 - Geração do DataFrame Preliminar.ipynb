# Importação das bibliotecas:

```python
import json
import glob
import pandas
```

# Listagem dos arquivos JSON originais:

```python
listaArquivos = glob.glob('*.json')
```

# Geração do Dataframe Preliminar:

```python
dfPreliminar = pandas.DataFrame()

for i in range (len(listaArquivos)):
    
    arquivo = listaArquivos[i]

    with open(arquivo) as f:
        
        dfTemp = json.load(f)
    
    dfTemp = pandas.json_normalize(dfTemp)
    
    dfPreliminar = dfPreliminar.append(dfTemp, ignore_index = True)
    
dfPreliminar.head()
```

# Exportação do Dataframe Preliminar em formato CSV:

```python
dfPreliminar.to_csv('dfPreliminar.csv', index = False)
```
